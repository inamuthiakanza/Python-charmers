# Puzzle 5 - current temperature of any city

Part 1: write a function called 'get_temperature' that takes a city name and looks up the current 
    temperature in celcius using the OpenWeatherMap API.
Part 2: make a list of these city names:
        - London
        - Beijing
        - Mumbai
        - Sydney
        - Singapore
        - Mogadishu
        - Cairo
        - Johannesburg

```python
cities = {'London', 'Beijing', 'Mumbai', 'Sydney', 'Singapore', 'Mogadishu', 'Cairo', 'Johannesburg'}
```

```python
def get_temperature (city):
    url = 'https://api.openweathermap.org/data/2.5/weather'
    params = {'APPID': key,
         'q': city,
         'units': 'metric'}
    response = requests.get(url, params)
    response.text
    weather = response.json()
    return(weather['main']['temp'])
```

```python
get_temperature('Jakarta')
```

```python
for name in cities:
    temp = get_temperature (name)
    print(name,temp)
```

```python
get_temperature('Anchorage')
```
