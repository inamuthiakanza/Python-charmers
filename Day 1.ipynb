# Puzzle 2 - word cloud
puzzle
make a word cloud of this article :
http://bit.ly/GIT_article_bahasa


```python
import requests
```

```python
response = requests.get('http://bit.ly/GIT_article_bahasa')
response.text
```

```python
%matplotlib inline
```

```python
import os

from os import path
from wordcloud import WordCloud

# Read the whole text.
text = response.text

# Generate a word cloud image
wordcloud = WordCloud().generate(text)

# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

# lower max_font_size
wordcloud = WordCloud(max_font_size=40).generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

# Quick eercise : Converting temperatures

$F = \frac{9}{5} * C + 32$

write a phyton function called 'centigrade_to_fahrenheit'. if it is currently 30 degrees outside, what is the temperature in fahrentheit?
    

```python
def centigrade_to_fahrenheit(temperature):
    fahrenheit = 9 / 5 * temperature + 32
    return fahrenheit
```

```python
centigrade_to_fahrenheit(30)
```

# Function version

```python
def funds_at_age(age, annual_investment=10_000_000):
    if age <= 20:
        return 0
    else:
        return(annual_investment +
              funds_at_age(age-1, annual_investment)
              ) * normalvariate(1.07, 0.05)
```

```python
funds_at_age(20)
```

```python
funds_at_age(21)
```

```python
funds_at_age(65)
```
